**Projeto 1: Calculadora de Variáveis Simples.**

```python
import pandas as pd

url_iris = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

df = pd.read_csv(url_iris)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['species'].value_counts()
```

```python
# A condição: "df na coluna 'species' é igual a 'setosa'?"
condicao = df['species'] == 'setosa'

# O filtro: Aplica a condição ao DataFrame inteiro
df_setosa = df[condicao]

df_setosa.shape
df_setosa.head()
```

```python
df_setosa['sepal_length'].mean()
```

```python
#df.groupby('coluna_de_interesse')['coluna_a_agregar'].funcao_de_agregação()

df.groupby('species')['sepal_length'].mean()
```

```python
import seaborn as sns

sns.histplot(data = df, x = 'sepal_length')
```

```python
sns.boxplot(data = df, x = 'species', y = 'sepal_length')
```

```python
sns.scatterplot(data = df, x = 'petal_length', y = 'petal_width', hue='species')
```
